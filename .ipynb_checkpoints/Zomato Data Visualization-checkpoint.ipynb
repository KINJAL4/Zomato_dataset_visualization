### importing necessary library

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import warnings
warnings.filterwarnings('ignore')
```

### Read dataset

```python
data=pd.read_csv("zomato.csv")
```

```python
data.head()
```

### as we are not going to use url,phone,dish_liked ,menu_item ..let's drop it.

```python
data.drop(['url','phone','menu_item'], axis=1,inplace=True)
```

```python
data.columns
```

### Rename columns with simpler names for use.

```python
data= data.rename(columns={'rate':'rating','approx_cost(for two people)':'cost','listed_in(city)':'city','listed_in(type)':'type'})
```

### Count duplicates in a DataFrame

```python
data.duplicated().sum()
```

### drop duplicates from original dataframe

```python
data.drop_duplicates(inplace=True)
```

```python
data.shape
```

### let's check if there any null value

```python
data.isna().sum()
```

### drop null values if there is any 

```python
data.dropna(how='any',inplace=True)
```

```python
data.shape
```

### let's check unique value of cost

```python
data['cost'].unique()
```

###  Convert data type of  cost from character to numeric

```python
data['cost']= data['cost'].astype(str)
data['cost']=data['cost'].apply(lambda x: x.replace(',',''))
data['cost']= data['cost'].astype(float)
```

```python
data.cost.unique()
```

```python
data.cost.value_counts()
```

### Let's check unique value of rating

```python
data['rating'].unique()
```

### Removing slash and convert data type of rating into numeric

```python
data= data.loc[data['rating']!='NEW']
data=data.loc[data['rating']!='-'].reset_index(drop=True)
remove_slash= lambda x: x.replace('/5','') if type(x)==np.str else x
data['rating'] = data['rating'].apply(remove_slash).str.strip().astype(float)
data['rating']
```

```python
data['rating'].dtype
```

```python
data.head()
```

### let's make the first word of name column's values  in upper case and  replace value of online_order and book_table 'yes' with 'True' and 'No' with 'False'.

```python
data['name']= data['name'].apply(lambda x: x.title())
data['online_order'].replace(('Yes','No'),(True,False),inplace=True)
data['book_table'].replace(('Yes','No'),(True,False),inplace=True)
```

```python
data.head()
```

### Encode the object into numeric categorical variable.

```python
def encode(data):
    for column in data.columns[~data.columns.isin(['rating','cost','votes'])]:
        data[column]= data[column].factorize()[0]
    return data
data_en= encode(data.copy())
data_en.head()
```

### Let's have a look on correlation between one variable with others using heatmap.

```python
plt.figure(figsize=[12,8])
sns.heatmap(data_en.corr(), annot=True,cmap='viridis')
plt.show()
```
